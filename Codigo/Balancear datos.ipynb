```python
import pandas as pd
import numpy   as np
data=pd.read_excel("../Datos/base proyecto gestión de cartera_210424.xlsx")
data['Mora'] =["Si" if data['Cuo. Ven.'][i]!= 0 else "No" for i in range(145) ]
data
```

```python
from sklearn.utils import resample
mora_0=data[data['Mora']=="Si"]
mora_1=data[data['Mora']=="No"]
mora_0
```

```python
mora_0_sobremuestro=resample(mora_0,replace=True,n_samples=len(mora_1),random_state=1)
mora_0_sobremuestro
```

```python
datos_balanceados=pd.concat([mora_0_sobremuestro,mora_1])
```

```python
print(datos_balanceados['Mora'].value_counts())
```

```python
datos_balanceados.reset_index(inplace=True,drop=True)
```

```python
datos_balanceados.to_csv('../Datos/DatosBalanceados.csv')
```
